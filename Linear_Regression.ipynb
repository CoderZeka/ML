```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
slope,intercept,r,p,std_err=stats.linregress(x,y)

y_new= lambda x:slope*x+intercept
plt.plot(x,y_new(x))
plt.scatter(x,y)
a=plt.plot(4,y_new(4),'x',color="red")
print(y_new(4))
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
